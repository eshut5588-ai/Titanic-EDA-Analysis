```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.drop(columns='Cabin', inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
df.groupby('AgeGroup')['Survived'].mean()
```

```python
df.groupby('Embarked')['Survived'].mean()
```

```python
df.groupby('FamilySize')['Survived'].mean()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title("Survival Rate by Family Size")
plt.show()
```

Children had higher survival rate compared to adults.

Passengers from certain embarkation ports had better survival chances.

Small families (1–3 members) had higher survival rates than very large families.

Passenger class negatively correlated with survival.

Fare positively correlated with survival.

Survival on the Titanic was influenced by multiple social and demographic factors. Age, family size, passenger class, and embarkation port played significant roles. Exploratory Data Analysis helped uncover hidden patterns and relationships within the dataset.
